# Data pre-processing/cleansing

```python
import pandas as pd

dataset = pd.read_csv("general_data.csv")

dataset.head()
```

```python
dataset.count()

dataset1 = dataset.dropna() # removing null records
```

```python
dataset1.head()
```

```python
dataset1['Attrition'] = dataset1['Attrition'].replace(['No'],'0') # Replacing attrition variable with 'Yes' as 1 & 'No' as 2

dataset1['Attrition'] = dataset1['Attrition'].replace(['Yes'],'1')
```

```python
#converting data type 'Attrition' variable to int 

dict = {'Attrition' : int}

dataset1 = dataset1.astype(dict) 
```

# Correlation - Hypothesis testing

Test statement 1 

H0 = There is significant correlation between age & attrition rate.

H1 = There is NO significant correlation between age & attrition rate.

```python
from scipy.stats import pearsonr

sampdf = dataset1.sample(n=100) #Getting random sample of 100 rows

stats,p = pearsonr (sampdf.Age,sampdf.Attrition)

print(stats,p)

if p < 0.5 :
    print('Since p value < 0.5 H0 null hypothesis is rejected & H1 is accepted')
else:
    print('Since p value > 0.5 H0 null hypothesis is accepted')
```
